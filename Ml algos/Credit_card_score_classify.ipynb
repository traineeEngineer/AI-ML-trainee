```python
from sklearn.preprocessing import LabelEncoder
cat_col=['Gender','Education','Marital Status','Home Ownership','Credit Score']

le=LabelEncoder()

for col in cat_col:
    df[col]=le.fit_transform(df[col])
```

```python
X=df.drop('Credit Score',axis=1)
Y=df['Credit Score']
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
lgr = LogisticRegression()
rf = RandomForestClassifier()
xgb = XGBClassifier()
```

```python
lgr_param={'C':[0.1,1,10]}
rf_param={'n_estimators':[100,200,300],'max_depth':[None,5,10]}
xgb_params={'learninh_rate':[0.1,0.01,0.001],'max_depth':[3,5,7]}
```

```python
lgr_grid_search=GridSearchCV(lgr,param_grid=lgr_param,cv=5,scoring='accuracy')
rf_grid_search=GridSearchCV(rf,param_grid=rf_param,cv=5,scoring='accuracy')
xgb_grid_search=GridSearchCV(xgb,param_grid=xgb_params,cv=7,scoring='accuracy')  
```

```python
lgr_grid_search.fit(X_train,Y_train)
rf_grid_search.fit(X_train,Y_train)
xgb_grid_search.fit(X_train,Y_train)
```

```python
print("Logistic Regression - Best Parameters:", lgr_grid_search.best_params_)
print("Logistic Regression - Best Score:", lgr_grid_search.best_score_)
print("Random Forest - Best Parameters:", rf_grid_search.best_params_)
print("Random Forest - Best Score:", rf_grid_search.best_score_)
print("XGBoost - Best Parameters:", xgb_grid_search.best_params_)
print("XGBoost - Best Score:", xgb_grid_search.best_score_)
```

```python
lgr_best_model=lgr_grid_search.best_estimator_
lgr_predict=lgr_best_model.predict(X_test)

rf_best_model=rf_grid_search.best_estimator_
rf_predict=rf_best_model.predict(X_test)

xgb_best_model=xgb_grid_search.best_estimator_
xgb_predict=xgb_best_model.predict(X_test)
```

```python
lgr_score=accuracy_score(Y_test,lgr_predict)
lgr_precision=precision_score(Y_test,lgr_predict,average='weighted')
lgr_recall=recall_score(Y_test,lgr_predict,average='weighted')
lgr_f1=f1_score(Y_test,lgr_predict,average='weighted')
print('='*30)

rf_score=accuracy_score(Y_test,rf_predict)
rf_precision=precision_score(Y_test,rf_predict,average='weighted')
rf_recall=recall_score(Y_test,rf_predict,average='weighted')
rf_f1=f1_score(Y_test,rf_predict,average='weighted')
print('='*30)

xgb_score=accuracy_score(Y_test,xgb_predict)
xgb_precision=precision_score(Y_test,xgb_predict,average='weighted')
xgb_recall=recall_score(Y_test,xgb_predict,average='weighted')
xgb_f1=f1_score(Y_test,xgb_predict,average='weighted')
```

```python
scores={
    'Model':['LogisticRegression','RandomForest','XGBoost'],
    'Accuracy':[lgr_score,rf_score,xgb_score],
    'Precision':[lgr_precision,rf_precision,xgb_precision],
    'Recall':[lgr_recall,rf_recall,xgb_recall],
    'F1_score':[lgr_f1,rf_f1,xgb_f1]
}
```
