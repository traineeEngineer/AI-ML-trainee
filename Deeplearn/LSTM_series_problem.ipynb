# Time ReSampling

```python
df.resample(rule='A').min()
```

```python
df.resample(rule='A').max()
```

```python
df.resample(rule='A').max()['Open'].plot() # rule A is year end frequency
```

```python
df.resample(rule='QS').max()
```

```python
df.resample(rule='QS').max()['Open'].plot()
```

```python
df.resample(rule='BQS').max().plot()
```

```python
df['Open'].resample(rule='BA').mean().plot(kind='bar')
```

```python
df['High'].rolling(10).mean()
```

```python
# Simple Moving Average :use smoothen the graph plots
df['Open:10 days rolling']=df['Open'].rolling(window=10,min_periods=1).mean()# min_period represent to aviod nan value based on window value
```

```python
df[['Open','Open:10 days rolling']].plot()
```

```python
#Cummaltive Moving Average :calculate average on both data like X1+X2,X1+X2+X3....
#Expanding
df['Open'].expanding().mean().plot(figsize=(10,5))
```

```python
#Exponential Moving Average ((anycol-prevEMA)*multiplier)+prevEMA) we dont have PrevEMA so initiate with SMA and consider as prevEMA.multiplier formaula :(2/windowsize+1)
df['EMA_1.0']=df['Open'].ewm(alpha=0.1,adjust=False).mean().plot(figsize=(12,6))
```

```python
#Exponential Weighted Moving Average
df['EMA_1.0']=df['Open'].ewm(span=5,adjust=False).mean().plot(figsize=(12,6))
```
