```python
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
from nltk.stem import WordNetLemmatizer
lemiter=WordNetLemmatizer()
lemiter.lemmatize(sentense)
```

```python
stemmer=PorterStemmer()
```

```python
#steming
for i in sentense:
    words=nltk.word_tokenize(i)
    for word in words:
        if word not in set(stopwords.words('english')):
            print(stemmer.stem(word))
```

```python
#lemmitization
for i in sentense:
    words=nltk.word_tokenize(i)
    for word in words:
        if word not in set(stopwords.words('english')):
            print(lemiter.lemmatize(word))
```

```python
#Bag of words
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(binary=True,ngram_range=(3,3))
X=cv.fit_transform(sentense)
cv.vocabulary_
```

```python
X[0].toarray()
```

```python
#apply stop words
import re
corpus=[]
for i in range(len(sentense)):
    text=re.sub('[^A-Za-z]','',sentense[i])
    text=text.lower()
    text=text.split()
    text=[lemiter.lemmatize(word) for word in text if not word in set(stopwords.words('english'))]
    text=''.join(text)
    corpus.append(text)
```

```python
#TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
cv=TfidfVectorizer(ngram_range=(2,3),max_features=10)
X=cv.fit_transform(sentense)
X[0].toarray()
```

```python
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good']
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
voc_size=10000
```

```python
onehot=[one_hot(words,voc_size) for words in sent]
```

```python
print(onehot,'\t')
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length=8
embded_docs=pad_sequences(onehot,padding='pre',maxlen=sent_length)
print(embded_docs)
```

```python
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(embded_docs))
```
