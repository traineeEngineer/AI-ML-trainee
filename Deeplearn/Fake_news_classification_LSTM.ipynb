```python
import pandas as pd
```

```python
df=pd.read_csv('../Deeplearning/Datasets/fake_train.csv')
```

```python
df.head()
```

```python
df.drop(['id','author'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
X=df.drop('label',axis=1)
Y=df['label']
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
voc_size=5000
```

```python
messages=X.copy()
```

```python
messages.shape
```

```python
messages.reset_index(inplace=True)
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
corpus=[]
for i in range(0,len(messages)):
    text=re.sub('[^a-zA-Z]',' ',messages['title'][i])
    text=text.lower().split()
    text=[ps.stem(word) for word in text if not word in stopwords.words('english')]
    text=' '.join(text)
    corpus.append(text)
```

```python
corpus[1]
```

```python
oh=[one_hot(words,voc_size) for words in corpus]
```

```python
oh
```

```python
sen_len=20
emb_doc=pad_sequences(oh,padding='pre',maxlen=sen_len) 
emb_doc 
```

```python
emb_doc[0]
```

```python
vec_feature=40
model=Sequential()
model.add(Embedding(voc_size,vec_feature,input_length=sen_len))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
import numpy as np
X_final=np.array(emb_doc)
Y_final=np.array(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X_final,Y_final,test_size=0.33)
```

```python
model.fit(X_train,Y_train,validation_data=(X_test,Y_test),epochs=10,batch_size=64)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test,y_pred)
```
