```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('../Deeplearning/Datasets/onlinefraud.csv')
```

```python
df.head()
```

```python
df.drop(['nameOrig','nameDest'],axis=1,inplace=True)
```

```python
df=df.rename(columns={'type':'Type'})
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
lbl_encode=LabelEncoder()
df['Type']=lbl_encode.fit_transform(df['Type'])
```

```python
df
```

```python
X=df.drop('isFraud',axis=1)
Y=df['isFraud']
```

```python
X
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.33,random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import LeakyReLU,PReLU,ReLU
```

```python
classfier=Sequential()
```

```python
#input layer creation
classfier.add(Dense(units=9,activation='relu'))
```

```python
#hidden layer creation
classfier.add(Dense(units=5,activation='relu'))
classfier.add(Dropout(0.2))
```

```python
#hiiden layer 2
classfier.add(Dense(units=3,activation='relu'))
classfier.add(Dropout(0.3))
```

```python
#output layer 
classfier.add(Dense(1,activation='sigmoid'))
```

```python
classfier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
import tensorflow 
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
```

```python
import tensorflow as tf
early_stops=tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    min_delta=0.02,
    patience=30,
    verbose=1,
    mode='auto',
    baseline=None,
    restore_best_weights=False
)
```

```python
model_hst=classfier.fit(X_train,Y_train,validation_split=0.33,batch_size=1000,epochs=500,callbacks=early_stops)
```

```python
model_hst.history.keys() 
```

```python
y_pred=classfier.predict(X_test)
y_pred=(y_pred>=0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,y_pred)
```

```python
cm
```

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(Y_test,y_pred)
```
