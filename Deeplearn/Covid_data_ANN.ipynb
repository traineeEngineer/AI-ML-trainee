```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('../Deeplearning/Datasets/Covid Data.csv')
```

```python
df['CLASIFFICATION_FINAL'] = df['CLASIFFICATION_FINAL'].replace([1,2,3], 1)
df['CLASIFFICATION_FINAL'] = df['CLASIFFICATION_FINAL'].replace([4,5,6,7], 0)
```

```python
df.head()
```

```python
df.drop(['DATE_DIED'],axis=1,inplace=True)
```

```python
df
```

```python
df.head()
```

```python
X=df.drop('CLASIFFICATION_FINAL',axis=1)
Y=df['CLASIFFICATION_FINAL']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=32)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
df.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import ReLU,PReLU
```

```python
classif=Sequential()
```

```python
classif.add(Dense(units=20,activation=PReLU()))
```

```python
classif.add(Dense(units=14,activation=ReLU()))
classif.add(Dropout(0.33))
```

```python
classif.add(Dense(units=11,activation=ReLU()))
classif.add(Dropout(0.2))
```

```python
classif.add(Dense(units=6,activation=PReLU()))
classif.add(Dropout(0.25))
```

```python
classif.add(Dense(1,activation='sigmoid'))
```

```python
classif.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
import tensorflow 
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
```

```python
import tensorflow as tf
early_stops=tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    min_delta=0.02,
    patience=40,
    verbose=1,
    mode='auto',
    baseline=None,
    restore_best_weights=True
)
```

```python
model_his=classif.fit(X_train,Y_train,batch_size=5000,validation_split=0.22,epochs=200,callbacks=early_stops)
```

```python
model_his.history.keys()
```

```python
y_pred=classif.predict(X_test)
y_pred=(y_pred>=0.4)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(Y_test,y_pred)
ac
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB,MultinomialNB
from xgboost import XGBClassifier
from sklearn.metrics import classification_report,accuracy_score
```

```python
lr=LogisticRegression()
lr.fit(X_train,Y_train)
y_pred=lr.predict(X_test)

print(accuracy_score(Y_test,y_pred))
print(classification_report(Y_test,y_pred))
```

```python
rf=RandomForestClassifier()
rf.fit(X_train,Y_train)
y_pred=rf.predict(X_test)

print(accuracy_score(Y_test,y_pred))
print(classification_report(Y_test,y_pred))
```

```python
gb=GaussianNB()
gb.fit(X_train,Y_train)
y_pred=gb.predict(X_test)

print(accuracy_score(Y_test,y_pred))
print(classification_report(Y_test,y_pred))
```

```python
xg=XGBClassifier()
xg.fit(X_train,Y_train)
y_pred=xg.predict(X_test)

print(accuracy_score(Y_test,y_pred))
print(classification_report(Y_test,y_pred))
```
